# Student's t distribution and Fat Tail behavior

```python
np.random.seed(42)
```

```python
plt.figure(figsize=(12,8))
x = np.linspace(-5,5,num=10000)
plt.plot(x, st.t.pdf(x,df=1), label='df=1')
plt.plot(x, st.t.pdf(x,df=3), label='df=3')
plt.plot(x, st.t.pdf(x,df=10), label='df=10')
plt.plot(x, st.norm.pdf(x), color='darkblue', label='gaussian, df=inf')
plt.legend(fontsize='large');
plt.axvline(2, color='r');
```

```python
print('Probability of 2 or extreme\n')
print('blue',1-st.t.cdf(2,df=1),sep='\t')
print('orange',1-st.t.cdf(2,df=3),sep='\t')
print('green',1-st.t.cdf(2,df=10),sep='\t')
# print(st.t.cdf(2,df=math.inf))
print('dark',1-st.norm.cdf(2),sep='\t')
```
