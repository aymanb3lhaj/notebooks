# Normalizing Data

In this notebook, we are trying to take a look and visualize most the normalization techniques.

* $\frac{\sigma}{\mu}$ : Removing the effect of different $\mu$ to compare $\sigma$.

```python
np.random.seed(100)
a = np.random.normal(10,10,size=10000)
b = np.random.normal(70,14,size=10000)
```

```python
sns.distplot(a);
sns.distplot(b);
```

```python
def norm(mu,sigma):
    return sigma/mu
```

```python
norm(10,10), norm(14,70)
```

```python
ap = np.random.normal(0,1,size=10000)
bp = np.random.normal(0,5.0,size=10000)
sns.distplot(ap);
sns.distplot(bp);
```
