```python
import warnings
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
np.random.seed(32)
plt.figure(figsize=(12,8))
sns.distplot(np.random.normal(0,1,size=10000),label='Null Hypothesis');
sns.distplot(np.random.normal(3,1,size=10000),label='Alternative Hypothesis',color='r');
plt.axvline(1.960,color='k'); # 95% confidence interval
plt.legend();
```
